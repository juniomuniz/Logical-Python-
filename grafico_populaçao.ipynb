```python
from matplotlib import pyplot as plt
anos = [ 1965, 1970, 1975, 1980, 1990, 1995, 2000, 2000]
populaçao = [199.76, 206.47, 209.88, 300.33, 388.64, 400.00, 444.44, 470.99]
x = [i + 0.1  for i, _ in enumerate(anos) ]

plt.bar(x, populaçao)

plt.ylabel("em milhares de pessoas")
plt.title("variaçao da populaçao de santos pos anos * 1000 ")
plt.xticks([i + 0.5 for i, _ in enumerate(anos )], anos)
plt.show
```

```python
import math 
from matplotlib pyplot as plt
def equaçao_2grau(a,b,c):
  delta = pow(b, 2 ) - 4*a*c 
  if delta < 0:
    return "raiz negativa, é imaginaria"
  else:
    raiz_delta = math.sqrt(delta)
    x1 = (-b + raiz_delta)/2* a 
    x2 =(-b - raiz_delta)/2*a 
    resultado = x1, x2
    return resultado

def parabola(a,b ,c):
  y = []
  for x in range(-2, 8):
    y.append(a*pow(x,2)+ b*x +c)
  x =[x for x in range(-2, 8)]
  plt.plot(x,y, color = 'purple', marker = '*', linestyle= 'solid')
  plt.grid(b = None, which= 'major' , axis ='both')
  plt.title('equaçao de 2 grau')
```
